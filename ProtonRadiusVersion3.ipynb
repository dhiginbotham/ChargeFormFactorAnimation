```python
"""
========================
Visualizing named colors
========================

Simple plot example with the named colors and its visual representation.
"""
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5),dpi=100)

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

```python
from lmfit import Model

list=[0.80,0.81,0.82,0.83,0.84,0.85,0.86,0.87,0.88,0.89,0.90,0.91,0.92]

q2=np.linspace(0.003,0.001,1000)

def linear(q2,b,m):
    return b+q2*m

rs=[];ms=[]

model=Model(linear)
for m in range(13):
    n=12-m
    data=(-6*np.log(GetAnyGE(q2,n))/q2/25.7)
    result=model.fit(data,q2=q2,b=0.84,m=0)
    #print(list[n],',',result.best_values['m']) 
    rs.append(list[n]) 
    ms.append(result.best_values['m'])

rs2=[];ms2=[]
q22=np.linspace(0.01,0.02,1000)

model2=Model(linear)
for m in range(13):
    n=12-m
    data2=(-6*np.log(GetAnyGE(q22,n))/q22/25.7)
    result2=model2.fit(data2,q2=q22,b=0,m=0)
    #print(list[n],',',result.best_values['m']) 
    rs2.append(list[n]) 
    ms2.append(result2.best_values['m'])
    #print(result2.fit_report())
    
rs=np.asarray(rs)
ms=np.asarray(ms)
rs2=np.asarray(rs2)
ms2=np.asarray(ms2)

plt.figure(dpi=150) 
plt.plot(ms,rs,'o',color='black',ms=5)
plt.plot(ms2,rs2,'o',color='red',ms=5)
result=model.fit(rs,q2=ms,b=0,m=0)
result2=model2.fit(rs2,q2=ms2,b=0,m=0)
plt.plot(ms,linear(ms,**result.best_values),color='black')
plt.plot(ms2,linear(ms2,**result2.best_values),'--',color='red')
plt.annotate('Intercept = 0.8406 fm +/- 0.03%',xy=(-1.5,0.910),fontsize=10)
plt.annotate('Intercept = 0.8288 fm +/- 0.03%',xy=(-1.5,0.920),fontsize=10,color='red')
plt.annotate('Slope Fits $Q^2$ < 0.01',xy=(-4,0.84),color='black')
plt.annotate('Slope Fits 0.01 < $Q^2$ < 0.02',xy=(-4,0.85),color='red')
plt.xlabel('Fitted Slope')
plt.ylabel('Radius [fm]')
plt.ylim(0.79,0.93)
plt.savefig('Slopes.png')
plt.show()

print(result.fit_report())
print(result2.fit_report())
```
